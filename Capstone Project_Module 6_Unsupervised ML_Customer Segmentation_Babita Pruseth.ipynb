## 3. ***Data Wrangling***

#### **3.1 Create TotalPrice Column**

````python
# Create TotalPrice Column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
df.columns
````

#### **3.2 Extract Year and Month from InvoiceDate**

````python
# Extract Year and Month from InvoiceDate
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df.columns
````

#### **3.3 Top Products by Quantity Sold**

````python
# Group by StockCode and Description to get the total quantity sold for each product
top_products_by_quantity = df.groupby(['StockCode', 'Description'])['Quantity'].sum().sort_values(ascending=False).head(10)

# Convert the result to a DataFrame for easier plotting
top_products_df = top_products_by_quantity.reset_index()
top_products_df.columns = ['StockCode', 'Product', 'Quantity Sold']

# Display the top-products by quantity sold DataFrame
print("Top Products by Quantity Sold")
top_products_df
````

#### **3.4 Top Selling Products**

````python
df['TotalSales'] = df['Quantity'] * df['UnitPrice']

# Group by StockCode and Description, then sum the TotalSales
top_products = df.groupby(['StockCode', 'Description'])['TotalSales'].sum().sort_values(ascending=False).head(10)

# Convert the resulting Series to a DataFrame
top_products_df = top_products.reset_index()

# Rename the columns for clarity
top_products_df.columns = ['StockCode', 'Description', 'Total Sales']

# Display the top-selling products DataFrame
print("Top Selling Products in the Dataset ")
top_products_df
````

#### **3.5 Least Selling Products**

````python
# df['TotalSales'] = df['Quantity'] * df['UnitPrice']

# Get the least selling products
least_products = df.groupby(['StockCode', 'Description'])['TotalSales'].sum().sort_values(ascending=True).head(10)

# Convert to a DataFrame
least_products_df = least_products.reset_index()
least_products_df.columns = ['StockCode', 'Product', 'Total Sales']
print("Top 10 Least Selling Products in the Dataset ")
least_products_df
````

#### **3.6 Group Data by CustomerID and InvoiceNo**

````python
# Group Data by CustomerID and InvoiceNo
customer_data = df.groupby(['CustomerID', 'InvoiceNo']).agg({
    'Quantity': 'sum',
    'TotalPrice': 'sum'
}).reset_index()
customer_data
````

#### **3.7 Standardize Country Names**

````python
# Standardize Country Names
df['Country'] = df['Country'].replace({
    'United States': 'USA',
    'United Kingdom': 'UK',
    'EIRE': 'Ireland'
})
df['Country']
````

#### **3.8 Top Countries by Sales**

````python
# Top N Countries by Sales
top_countries = df.groupby('Country')['TotalPrice'].sum().nlargest(5).index
df_top_countries = df[df['Country'].isin(top_countries)]
top_countries
#df_top_countries
````

#### **3.9 Months Generating the Highest Business**

````python
# Ensure 'InvoiceDate' is in datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Extract month and year from InvoiceDate
df['Month'] = df['InvoiceDate'].dt.month_name()
df['Year'] = df['InvoiceDate'].dt.year

# Calculate total sales per month
monthly_sales = df.groupby(['Year', 'Month'])['TotalPrice'].sum().reset_index()

# Sort months by sales
top_months = monthly_sales.sort_values(by='TotalPrice', ascending=False).head(5)
print(top_months)
````

#### **4.0 Most Popular Purchasing Days**

````python
# Extract day of the week from InvoiceDate
df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()

# Calculate total sales per day of the week
day_sales = df.groupby('DayOfWeek')['TotalPrice'].sum().reset_index()

# Sort days by sales
top_days = day_sales.sort_values(by='TotalPrice', ascending=False)
print(top_days)
````

#### **4.1 Most Customer Purchases by Time of Day**

````python
# Extract hour from InvoiceDate
df['Hour'] = df['InvoiceDate'].dt.hour

# Calculate total sales per hour
hourly_sales = df.groupby('Hour')['TotalPrice'].sum().reset_index()

# Find peak purchasing hours
peak_hours = hourly_sales.sort_values(by='TotalPrice', ascending=False)
print(peak_hours)
````

#### **4.2 Peak Purchasing Times**

````python
# Create a new column for time of day
def categorize_time(hour):
    if 10 <= hour < 14:
        return 'Midday'
    elif 0 <= hour < 10:
        return 'Morning'
    else:
        return 'Evening'

df['TimeOfDay'] = df['Hour'].apply(categorize_time)

# Calculate total sales per time of day
time_of_day_sales = df.groupby('TimeOfDay')['TotalPrice'].sum().reset_index()

# Sort by sales
peak_times = time_of_day_sales.sort_values(by='TotalPrice', ascending=False)
print(peak_times)
````

## ***5. Hypothesis Testing***

Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

##### **Hypothetical Statement - 1**

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.



Null Hypothesis (H0): The average retention rate for customers acquired in December 2010 is equal to that of customers acquired in January 2011.

Alternative Hypothesis (H1): The average retention rate for customers acquired in December 2010 is different from that of customers acquired in January 2011.

````python
# import pandas as pd
# import numpy as np
from scipy import stats

# 1. Calculate the Retention Rate for each Cohort Month
cohort_data = df.groupby(['CohortMonth', 'CohortIndex']).agg({
    'CustomerID': pd.Series.nunique
}).reset_index()

# Rename the CustomerID column to reflect it represents unique customers
cohort_data.rename(columns={'CustomerID': 'TotalCustomers'}, inplace=True)

# Pivot the data to have CohortIndex as columns and CohortMonth as rows
cohort_pivot = cohort_data.pivot_table(index='CohortMonth', columns='CohortIndex', values='TotalCustomers')

# Calculate the retention rate
retention = cohort_pivot.divide(cohort_pivot.iloc[:, 0], axis=0) * 100

# 2. Extract retention rates for the cohorts of interest
retention_dec_2010 = retention.loc['2010-12'].dropna().values
retention_jan_2011 = retention.loc['2011-01'].dropna().values

##### Perform appropriate statistical test.

# 3. Perform an independent t-test
t_stat, p_value = stats.ttest_ind(retention_dec_2010, retention_jan_2011)

# 4. Display the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in retention rates.")
else:
    print("Fail to reject the null hypothesis: No significant difference in retention rates.")
````

##### Which statistical test have you done to obtain P-Value?

The statistical test performed to obtain the p-value is an independent two-sample t-test. This test is used to determine whether there is a statistically significant difference between the means of two independent groups—in this case, the retention rates of the cohorts from December 2010 and January 2011.

##### Why did you choose the specific statistical test?

The independent two-sample t-test was chosen because it is appropriate for comparing the means of two independent groups to determine if there is a statistically significant difference between them. In this case, the two groups are the retention rates of customers who first made a purchase in December 2010 and those who did so in January 2011. Since we are interested in comparing the mean retention rates between these two cohorts, the t-test is the suitable choice.

##### **Hypothetical Statement - 2**

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.



Null Hypothesis (H0): The retention rate does not decrease as the cohort index increases.

Alternative Hypothesis (H1): The retention rate decreases as the cohort index increases.


````python
# import pandas as pd
import statsmodels.api as sm

# Reset index
df_long = retention_matrix.reset_index()

# Melt the DataFrame
df_long = df_long.melt(id_vars='CohortMonth', var_name='CohortIndex', value_name='RetentionRate')


# Convert 'CohortIndex' and 'RetentionRate' to numeric, forcing errors to NaN
df_long['CohortIndex'] = pd.to_numeric(df_long['CohortIndex'], errors='coerce')
df_long['RetentionRate'] = pd.to_numeric(df_long['RetentionRate'], errors='coerce')

# Drop rows with missing values if any
df_long = df_long.dropna(subset=['CohortIndex', 'RetentionRate'])

# Prepare the data
X = df_long[['CohortIndex']]
X = sm.add_constant(X)  # Add an intercept
y = df_long['RetentionRate']

# Fit the linear regression model
model = sm.OLS(y, X).fit()

# Print the summary of the model
print(model.summary())

# Check significance
if model.pvalues['CohortIndex'] < 0.05:
    print("Reject the null hypothesis: Retention rate decreases with an increase in cohort index.")
else:
    print("Fail to reject the null hypothesis: No significant decrease in retention rate with an increase in cohort index.")
````

##### Which statistical test have you done to obtain P-Value?

The p-value is obtained from the t-test associated with the regression coefficient for the CohortIndex variable, using Ordinary Least Squares (OLS) linear regression analysis. In this context, With a p-value of 0.000, which is less than 0.05, null hypothesis is rejected and concluded that there is a significant decrease in the retention rate with an increase in the cohort index.

##### Why did you choose the specific statistical test?

The t-test is chosen because it evaluates whether the regression coefficient for CohortIndex is significantly different from zero. This helps determine if CohortIndex has a significant impact on the RetentionRate. In the context of linear regression, the t-test assesses the individual contribution of each predictor to the model.

#### **6.1.4 Customer segments with RFM Model**

Assign scores to customers based on recency, frequency, and monetary value.

The simplest way to create customers segments from RFM Model is to use **Quantiles**. We assign a score from 1 to 4 to Recency, Frequency and Monetary. Four is the best/highest value, and one is the lowest/worst value. A final RFM score is calculated simply by combining individual RFM score numbers.

Note: Quintiles (score from 1-5) offer better granularity, in case the business needs that but it will be more challenging to create segments since we will have 555 possible combinations. So, we will use quartiles.

##### **Calculate RFM Scores**

````python
# Define custom bins based on quantiles or fixed ranges
recency_bins = [0, rfm_df['Recency'].quantile(0.25), rfm_df['Recency'].quantile(0.50), rfm_df['Recency'].quantile(0.75), rfm_df['Recency'].max()]
frequency_bins = [0, rfm_df['Frequency'].quantile(0.25), rfm_df['Frequency'].quantile(0.50), rfm_df['Frequency'].quantile(0.75), rfm_df['Frequency'].max()]
monetary_bins = [0, rfm_df['MonetaryValue'].quantile(0.25), rfm_df['MonetaryValue'].quantile(0.50), rfm_df['MonetaryValue'].quantile(0.75), rfm_df['MonetaryValue'].max()]

# Create RecencyScore column using custom bins
rfm_df['RecencyScore'] = pd.cut(rfm_df['Recency'], bins=recency_bins, labels=list(range(4, 0, -1)), include_lowest=True)

# Create FrequencyScore column using custom bins
rfm_df['FrequencyScore'] = pd.cut(rfm_df['Frequency'], bins=frequency_bins, labels=list(range(1, 5)), include_lowest=True)

# Create MonetaryScore column using custom bins
rfm_df['MonetaryScore'] = pd.cut(rfm_df['MonetaryValue'], bins=monetary_bins, labels=list(range(1, 5)), include_lowest=True)

# Display first 5 rows of the dataframe with scores
#print("\nFirst 5 rows of rfm_df with scores:\n", rfm_df.head())

# Concatenate the RecencyScore, FrequencyScore, and MonetaryScore columns into one RFMScore column
rfm_df['RFMScore'] = rfm_df['RecencyScore'].astype(str) + rfm_df['FrequencyScore'].astype(str) + rfm_df['MonetaryScore'].astype(str)

# Display the updated dataframe with RFMScore
print("\nFirst 5 rows of rfm_df with RFMScore:\n", rfm_df.head())

# Sort the dataframe by MonetaryValue in descending order and reset the index
#rfm_df2 = rfm_df[rfm_df['RFMScore'] == '444'].sort_values('MonetaryValue', ascending = False)
#rfm_df2.head(10)
````

##### **Segment Customers**

````python
# Segment Customers
# import pandas as pd

# Define segments based on RFM scores
def segment_customers(row):
    if row['RecencyScore'] == 4 and row['FrequencyScore'] == 4 and row['MonetaryScore'] == 4:
        return 'Champions'
    elif row['FrequencyScore'] == 4 and row['MonetaryScore'] == 4:
        return 'Loyal Customers'
    elif row['RecencyScore'] == 4 and row['FrequencyScore'] == 4:
        return 'Potential Loyalists'
    elif row['RecencyScore'] == 4:
        return 'New Customers'
    elif row['RecencyScore'] == 1 and (row['FrequencyScore'] == 4 or row['MonetaryScore'] == 4):
        return 'At-Risk Customers'
    elif row['RecencyScore'] == 1:
        return 'Lost Customers'
    else:
        return 'Other'

# Apply segmentation
rfm_df['Segment'] = rfm_df.apply(segment_customers, axis=1)
rfm_df
````

4: Highest spend.
1: Lowest spend.

## ***7. ML Model Implementation***

#### **7.1. K-means Implementation**

Determining Number of Clusters: Choosing the right number of clusters for K-means can be challenging. Evaluating different cluster counts using silhouette coefficients can help assess clustering quality.

Silhouette Coefficient: Measures how well data points fit within their assigned cluster, ranging from -1 (poor fit) to 1 (good fit). Higher values indicate better clustering.

Elbow Method: Identifies the optimal number of clusters by plotting the number of clusters against the distortion or inertia. The "elbow" point on the plot suggests a suitable cluster count.

K-means++ Initialization: Improves K-means performance by selecting initial centroids more strategically, which enhances the chances of achieving better clustering results.

### **7.1.1 K-Means with silhouette_score**

````python
# Import necessary libraries
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# create an empty list silhouette_scores to store the silhouette scores.
silhouette_scores = []

# Loop over different values of K
for n_clusters in range(2, 16):
  # Initialize the K-Means model with number of clusters
  kmeans = KMeans(init = 'k-means++', n_clusters = n_clusters)

  # Fit the KMeans Model to the data
  kmeans.fit(X)

  # Predict the cluster labels for each data point
  labels = kmeans.labels_

  # Calculate the Silhouette score
  silhouette = silhouette_score(X, labels)

  # Append the silhouette_score to the array
  silhouette_scores.append(silhouette)

  # Print the silhouette score
  print(f'Silhouette score for {n_clusters} clusters: {silhouette:.3f}')
````

The silhouette score measures how similar each data point is to its own cluster compared to other clusters. A higher silhouette score indicates better-defined clusters. 

In this case, the score decreases as the number of clusters increases beyond 2, suggesting that adding more clusters might not improve clustering quality and could potentially make the clusters less distinct.

````python
# Plot the silhouette scores
plt.plot(range(2, 16), silhouette_scores, '-o', color='purple', markersize=10, linewidth=2)
plt.xlabel('Number of clusters (K)', fontsize=14)
plt.ylabel('Silhouette score', fontsize=14)
plt.title('Silhouette score for different values of K', fontsize=16)
plt.xticks(range(2, 16), fontsize=12)
plt.yticks(fontsize=12)
plt.show()
````

The plot of silhouette scores over different numbers of clusters showing that the best score is achieved with 2 clusters. This indicates that 2 is likely the optimal number of clusters for this data, as it provides the clearest separation and best clustering quality. 

##### **Visualization of Customer Segmentation Based on RFM Features**

#### **7.1.2 K-Means with Elbow method**

The Elbow Method plot is commonly used to identify the optimal number of clusters in a K-means clustering algorithm.

````python
# Input data X

# Initialize an empty list to store the WCSS values for different number of clusters
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)  # Create a KMeans instance for each number of clusters
    kmeans.fit(X)  # Fit the KMeans model to the input data X
    wcss.append(kmeans.inertia_)  # Append the WCSS value to the list for the current number of clusters

# Plot the WCSS values against the number of clusters
plt.figure(figsize=(10,6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1, 11, 1))
plt.grid(True)
plt.show()
````

It identified that k = 2 is the optimal number of clusters, it means that dividing this data into two clusters is the most appropriate choice.

#### **7.2 Agglomerative Hierarchial Clustering**

Agglomerative Hierarchical Clustering is a bottom-up technique that merges similar clusters until a specified number is reached. It relies on a distance measure between data points and can produce a dendrogram to visualize the clustering process. While it effectively captures hierarchical relationships, it's computationally intensive for large datasets and less suitable for high-dimensional data

````python
from sklearn.cluster import AgglomerativeClustering

# Create an instance of AgglomerativeClustering with 2 clusters and ward linkage
model = AgglomerativeClustering(n_clusters=2, linkage='ward')

# Fit the input data X to the model
model.fit(X)

# Get the cluster labels
labels = model.labels_
````

````python
# Import necessary libraries
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt

# Set the figure size and title for the dendrogram plot
plt.figure(figsize=(15, 12))
plt.title('Agglomerative Hierarchical Clustering Dendogram')

# Set the x and y-axis labels for the dendrogram plot
plt.xlabel('Sample index')
plt.ylabel('Distance')

# Create a linkage matrix using the input data X and the ward linkage method
Z = linkage(X, 'ward')

# Plot the dendrogram with specified parameters
dendrogram(Z, leaf_rotation=90.0, p=25, color_threshold=80, leaf_font_size=10, truncate_mode='level')

# Ensure tight layout of the plot
plt.tight_layout()
````


Here: 

X-Axis (Sample Index): Represents the index or position of data points in the dataset. Each sample starts as its own cluster at the bottom.

Y-Axis (Distance): Shows the distance or dissimilarity at which clusters merge. Higher values indicate greater dissimilarity between merged clusters.

The dendrogram illustrates how individual samples are grouped into clusters. Lower on the plot, individual samples are clustered together, and as we move up, these clusters merge at increasing levels of dissimilarity.

#### **7.3 DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

DBSCAN is a clustering algorithm that groups points based on density. It finds dense regions separated by less dense areas, does not require a predefined number of clusters, and is resilient to noise and outliers. Key parameters are the distance (Eps) and the minimum number of points (MinPts) to form clusters.

````python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

# Create and fit the DBSCAN model
dbscan = DBSCAN(eps=0.5, min_samples=15)
dbscan.fit(X)

# Plot the results
plt.scatter(X[:,0], X[:,1], c=dbscan.labels_, cmap='rainbow')
plt.title('DBSCAN Clustering')
plt.xlabel('Recency')
plt.ylabel('Frequency')
plt.show()
````

The scatter plot visualizes DBSCAN clustering results, with each point colored according to its cluster label.

X and Y Axis: Represent the two features of the dataset (e.g., Recency and Frequency). The scatter plot shows how points are grouped based on their feature values.

Clusters and Outliers: Points closer to each other form clusters, while outliers or noise are labeled as -1. The plot reveals the density and distribution of clusters, helping to identify patterns or anomalies in the data.
